<h2 style="color:#FF69B4">FINDING THE  NULL VALUE.

<h4>BY USING isnull()</h4>

```python
car.isnull().sum()
```

```python
#There are no null values. Say, there were 2 null values in 'cylinders' column. This is how we would remove it:
car['cylinders'].fillna(car['cylinders'].mean(),inplace= True)
```

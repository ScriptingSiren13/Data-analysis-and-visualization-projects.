<h2 style="color:#FF69B4">SHOW THE NUMBER OF VALUES FOR THE COLUMNS "CONFIRMED", "DEATHS" AND "RECOVERED" CASES IN EACH REGION</h2>

```python
data.head(2)
```

```python
#whenever we have to find 'each' value, use groupby()
# data.groupby('Region')
data.groupby('Region')[['Confirmed', 'Deaths', 'Recovered']].sum().head(3)
```

<h2 style="color:#FF69B4" >REMOVE ALL THE RECORDS WHERE CONFIRMED CASE IS LESS THAN 2000</h2>

```python
data.head(3)
```

```python
#TO check the dataframe where confirmed less than 2000
data[data['Confirmed'] <2000]
```

```python
#To remove thsoe recods
data[~(data['Confirmed'] <2000)]
```

<h2 style="color:#FF69B4">IN WHICH REGION, MAX NO OF CONFIRMED CASES WERE RECORDED</h2>

```python
data.head(3)
```

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=False)
```

<H2 style="color:#FF69B4">IN WHICH REGION, MINIMUM NUMBER OF DEATH CASES WERE RECORDED</H2>

```python
data.head(3)
```

```python
data.groupby('Region').Deaths.sum().sort_values(ascending=True)
```
